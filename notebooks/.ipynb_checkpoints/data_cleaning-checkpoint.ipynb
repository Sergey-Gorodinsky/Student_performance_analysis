Data Cleaning

1. Remove redundant columns;
2. Rename columns;
3. Check data types;
4. Remove duplicates;
5. Remove NaN values;
6. Check for outliers;
7. Check the dataset for logical errors;


```python
import pandas as pd
```

```python
path = r"/Users/sergeygorodinsky/Desktop/py/Student_Perfomance_Analysis/data/raw/student_learning_habits_dataset.csv"
original_data = pd.read_csv(path)

df = original_data.copy()
df
```

1. Remove redundant columns;


```python
df.columns
```

```python
columns_to_keep = ['daily_study_hours', 'online_courses_taken',
       'avg_sleep_hours', 'exam_score']
columns_to_drop = ['student_id']

df.drop(columns=columns_to_drop,inplace=True)
df.head(3)
```

2. Rename columns;


Column names were reviewed for readability. All columns follow the snake_case naming convention, so no renaming was required.

3. Check data types;

```python
df.dtypes
```

All columns were assigned appropriate data types. Numeric variables such as study hours, sleep hours, and exam scores were stored as int or float so no conversion was required.

4. Remove duplicates;

```python
df.duplicated().sum()
```

No duplicate rows were found, so no records were removed.

5. Remove NaN values;

```python
df.isna().sum()
```

No NaN values were found, so nothing was removed.

6. Check for outliers using IQR method;

```python
numeric_columns = ['daily_study_hours', 'online_courses_taken',
       'avg_sleep_hours', 'exam_score']

for col in numeric_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

outliers
```

The dataset was checked for outliers using the IQR method. No outliers were identified.

7. Check the dataset for logical errors;

```python
logical_errors = df[
    (df["daily_study_hours"] < 0) | (df["daily_study_hours"] > 24) |
    (df["online_courses_taken"]) < 0 |
    (df["avg_sleep_hours"] < 0) | (df["avg_sleep_hours"] > 24)
]
```

```python
logical_errors
```

No logical errors were found.
